```python
arr = np.random.randn(1000000)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# arr has 1,000,000 values, so a histogram is more useful than a line plot
axes[0].hist(arr, bins=50, color="steelblue", edgecolor="white")
axes[0].set_title("Distribution of arr")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Frequency")

# new_arr is small, so a simple line plot works well
axes[1].plot(new_arr, marker="o")
axes[1].set_title("Values in new_arr")
axes[1].set_xlabel("Index")
axes[1].set_ylabel("Value")

plt.tight_layout()
plt.show()
```
