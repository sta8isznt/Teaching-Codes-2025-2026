```python
students = pd.DataFrame({
    "student_id": [101, 102, 103, 104],
    "Name": ["Anna", "Nikos", "Maria", "John"],
    "Department": ["CS", "Math", "Physics", "CS"]
})

grades = pd.DataFrame({
    "id": [101, 102, 104, 105],
    "Score": [85, 90, 78, 88]
})
```

```python
students_with_grades = students.merge(grades, left_on="student_id", right_on="id", how="left")
students_with_grades
```

```python
students_with_grades[students_with_grades["Score"].isna()]
```

```python
students_with_grades["Score"] = students_with_grades["Score"].fillna(0)
students_with_grades
```

```python
students_with_grades["Passed"] = students_with_grades["Score"] >= 50
students_with_grades
```

Δίνονται:
students = pd.DataFrame({
    "student_id": [101, 102, 103, 104],
    "Name": ["Anna", "Nikos", "Maria", "John"],
    "Department": ["CS", "Math", "Physics", "CS"]

grades = pd.DataFrame({

    "Score": [85, 90, 78, 88]
})
Ζητούνται:
1.	Να γίνει left merge ώστε να κρατηθούν όλοι οι φοιτητές.
2.	Να εντοπιστούν ποιοι φοιτητές δεν έχουν βαθμό.
3.	Να αντικατασταθούν οι ελλείπουσες τιμές της Score με 0.
4.	Να προστεθεί στήλη Passed που να είναι True αν Score >= 50, αλλιώς False.


```python
students_with_grades = students.merge(grades, left_on="student_id", right_on="id", how="left")
missing_grades = students_with_grades[students_with_grades["Score"].isna()]
students_with_grades["Score"] = students_with_grades["Score"].fillna(0)
students_with_grades["Passed"] = students_with_grades["Score"] >= 50

missing_grades, students_with_grades
```
